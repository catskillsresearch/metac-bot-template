# Barrel of Monkeys

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define selectorate model parameters for each country
selectorate_coefficients = {
    'Israel': {'S': 0.7, 'W': 0.3, 'L': 0.2},
    'Russia': {'S': 0.4, 'W': 0.1, 'L': 0.5},
    'Ukraine': {'S': 0.6, 'W': 0.25, 'L': 0.15},
    'USA': {'S': 0.9, 'W': 0.6, 'L': 0.1},
    'Iran': {'S': 0.5, 'W': 0.2, 'L': 0.3}
}

# Initialize directed graph
G = nx.DiGraph()

# Add nodes with selectorate attributes
for country in selectorate_coefficients:
    G.add_node(country, **selectorate_coefficients[country])

# Define conflict relationships with weights and secondary effects
conflict_relationships = [
    ('Russia', 'Ukraine', {'type': 'military', 'weight': 0.9}),
    ('Ukraine', 'Russia', {'type': 'resistance', 'weight': 0.8}),
    ('Russia', 'USA', {'type': 'cyber', 'weight': 0.4}),
    ('Israel', 'USA', {'type': 'lobbying', 'weight': 0.7}),
    ('Ukraine', 'Israel', {'type': 'tech_coop', 'weight': 0.6}),
    ('Israel', 'Iran', {'type': 'military', 'weight': 0.8, 
                       'secondary': {'Ukraine': 0.3, 'Russia': -0.4}}),
    ('Iran', 'Israel', {'type': 'proxy', 'weight': 0.5, 
                       'secondary': {'USA': -0.2}})
]

G.add_edges_from(conflict_relationships)

# Add economic dependency parameters
economic_factors = {
    'Russia': {'energy_exports': 0.8, 'military_spending': 0.9},
    'Ukraine': {'aid_dependency': 0.7, 'agriculture': 0.6},
    'USA': {'military_aid': 0.5, 'diplomatic_leverage': 0.8},
    'Israel': {'tech_exports': 0.9, 'security_coop': 0.85},
    'Iran': {'oil_exports': 0.7, 'proxy_force': 0.6}
}

for node in G.nodes:
    G.nodes[node].update(economic_factors[node])

# Visualization parameters
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
node_sizes = [G.nodes[c]['W'] * 3000 for c in G.nodes]
edge_weights = [abs(G[u][v]['weight'])*4 for u,v in G.edges]
colors = {'military': 'red', 'proxy': 'orange', 'tech_coop': 'green',
          'cyber': 'blue', 'lobbying': 'purple', 'resistance': 'brown'}
edge_colors = [colors[G[u][v]['type']] for u,v in G.edges]

# Draw elements
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, 
                      width=edge_weights, arrowsize=25)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Add selectorate coefficients labels
coefficient_labels = {c: f"S={G.nodes[c]['S']}\nW={G.nodes[c]['W']}\nL={G.nodes[c]['L']}" 
                      for c in G.nodes}
nx.draw_networkx_labels(G, pos, labels=coefficient_labels, 
                        font_size=9, font_color='red', 
                        verticalalignment='bottom')

plt.title("Enhanced Selectorate Model: Conflict Network Analysis", pad=20)
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define selectorate model parameters
selectorate_coefficients = {
    'Israel': {'S': 0.7, 'W': 0.3, 'L': 0.2},
    'Russia': {'S': 0.4, 'W': 0.1, 'L': 0.5},
    'Ukraine': {'S': 0.6, 'W': 0.25, 'L': 0.15},
    'USA': {'S': 0.9, 'W': 0.6, 'L': 0.1},
    'Iran': {'S': 0.5, 'W': 0.2, 'L': 0.3}
}

G = nx.DiGraph()

# Add nodes with attributes
for country in selectorate_coefficients:
    G.add_node(country, **selectorate_coefficients[country])

# Define relationships with clear support/opposition
relationships = [
    # Support relationships (blue)
    ('Iran', 'Russia', {'type': 'support', 'weight': 0.7}),
    ('USA', 'Israel', {'type': 'support', 'weight': 0.8}),
    ('Israel', 'Ukraine', {'type': 'support', 'weight': 0.6}),
    
    # Opposition relationships (red)
    ('Russia', 'Ukraine', {'type': 'opposition', 'weight': 0.9}),
    ('Israel', 'Iran', {'type': 'opposition', 'weight': 0.85}),
    ('Iran', 'Israel', {'type': 'opposition', 'weight': 0.75}),
    
    # Additional interactions
    ('USA', 'Ukraine', {'type': 'support', 'weight': 0.7}),
    ('Russia', 'USA', {'type': 'opposition', 'weight': 0.4}),
    ('Ukraine', 'Russia', {'type': 'opposition', 'weight': 0.8})
]

G.add_edges_from(relationships)

# Visualization setup
plt.figure(figsize=(14, 10))
pos = nx.circular_layout(G)
node_sizes = [G.nodes[c]['W']*4000 for c in G.nodes]

# Color scheme
edge_colors = ['blue' if G[u][v]['type'] == 'support' else 'red' 
               for u,v in G.edges]
edge_widths = [G[u][v]['weight']*6 for u,v in G.edges]

# Draw elements
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='lightgray', edgecolors='black')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=edge_widths, 
                      arrowsize=25, connectionstyle='arc3,rad=0.1')
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Selectorate labels
nx.draw_networkx_labels(G, pos, 
                        labels={c: f"S={G.nodes[c]['S']}\nW={G.nodes[c]['W']}\nL={G.nodes[c]['L']}" 
                                for c in G.nodes},
                        font_size=9, font_color='blue', verticalalignment='bottom')

# Create legend
legend_patches = [
    mpatches.Patch(color='blue', label='Support Relationship'),
    mpatches.Patch(color='red', label='Opposition Relationship'),
    mpatches.Line2D([0], [0], marker='o', color='w', label='Node Size = Winning Coalition (W)',
                  markerfacecolor='gray', markersize=15)
]
plt.legend(handles=legend_patches, loc='upper right', framealpha=1)

plt.title("Geopolitical Relationships with Selectorate Model Parameters", pad=25)
plt.axis('off')
plt.tight_layout()
plt.show()
```
